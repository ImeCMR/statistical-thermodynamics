Plot the probability density for a normalized Gaussian Distribution centered at zero and with Ïƒ= 1

```python
# Normalized gaussian distribution with mean=0 and standard deviation=1

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate x values
x = np.linspace(-5, 5, 1000)

# Calculate the Gaussian distribution
y = norm.pdf(x, loc=0, scale=1)

# Plot the Gaussian distribution
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Normalized Gaussian Distribution (Mean=0, Standard Deviation=1)") # Changed variance to standard deviation in the title
plt.grid(True)
plt.show()
```

Plot the probability density for a normalized Gaussian Distribution centered at zero and with Ïƒ= 0.1, 0.01, 0.001

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate x values
x = np.linspace(-5, 5, 1000)
scales = [0.1, 0.01, 0.001]

# Loop through each scale and generate a plot
for scale in scales:
    # Calculate the Gaussian distribution
    y = norm.pdf(x, loc=0, scale=scale)
    
    # Plot the Gaussian distribution
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("Probability Density")
    plt.title(f"Gaussian Distribution (Mean=0, Standard Deviation={scale})")
    plt.grid(True)
    plt.show()
```
